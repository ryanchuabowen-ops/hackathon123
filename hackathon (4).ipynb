```python
###defunc
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# 1. Prepare Data & Train Model (Including Lags 1-4)
pt1_ml = pt1.copy()
for i in range(1, 5):
    pt1_ml[f'perf_lag_{i}'] = pt1_ml.groupby('student_id')['performance_index'].shift(i)

pt1_ml = pt1_ml.dropna()

features = [
    'study_hours', 'sleep_hours', 'stress_level', 
    'attendance_rate', 'learning_efficiency', 'fatigue_index',
    'perf_lag_1', 'perf_lag_2', 'perf_lag_3', 'perf_lag_4'
]
X = pt1_ml[features]
y = pt1_ml['performance_index']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
model = SVR(kernel='linear', C=1.0) 
model.fit(X_scaled, y)
weights = model.coef_[0]

# 2. Setup Subplots
fig_svm = make_subplots(
    rows=1, cols=2, 
    column_widths=[0.7, 0.3],
    subplot_titles=("Performance Projections & Baseline", "AI Impact Factors"),
    horizontal_spacing=0.12
)

student_list = pt1_ml['student_id'].unique()

# 3. Add Traces (4 traces per student: 3 lines, 1 bar)
for student in student_list:
    s_df = pt1_ml[pt1_ml['student_id'] == student]
    s_X_scaled = scaler.transform(s_df[features])
    
    y_pred = model.predict(s_X_scaled)
    
    # Simulation: Scenario improvement
    s_opt = s_df[features].copy()
    s_opt['sleep_hours'] += 1.5
    s_opt['fatigue_index'] = (s_opt['fatigue_index'] - 2).clip(0, 10)
    y_opt = model.predict(scaler.transform(s_opt))

    # --- Trend Chart (Column 1) ---
    # Actual Data
    fig_svm.add_trace(go.Scatter(
        x=s_df['week'], y=s_df['performance_index'],
        name="Actual History", mode='lines+markers', visible=False,
        line=dict(color='#1e293b', width=2), legendgroup="group1"
    ), row=1, col=1)

    # AI Baseline (Historical Fit)
    fig_svm.add_trace(go.Scatter(
        x=s_df['week'], y=y_pred,
        name="AI Baseline (Model Fit)", mode='lines', visible=False,
        line=dict(color='#3b82f6', dash='dash', width=2), legendgroup="group1"
    ), row=1, col=1)

    # Projected Potential
    fig_svm.add_trace(go.Scatter(
        x=s_df['week'], y=y_opt,
        name="Projected Potential", mode='lines', visible=False,
        line=dict(color='#10b981', width=4), legendgroup="group1"
    ), row=1, col=1)

    # --- Reasoning Bar Chart (Column 2) ---
    sorted_idx = np.argsort(weights)
    fig_svm.add_trace(go.Bar(
        x=weights[sorted_idx],
        y=[features[i] for i in sorted_idx],
        orientation='h',
        name="Feature Weight",
        visible=False,
        marker=dict(color=['#ef4444' if w < 0 else '#3b82f6' for w in weights[sorted_idx]]),
        showlegend=False # Keep legend focused on the line chart
    ), row=1, col=2)

# 4. Initialize Interactivity (4 traces per student)
for i in range(4):
    fig_svm.data[i].visible = True

buttons = []
for i, student in enumerate(student_list):
    visibility = [False] * len(fig.data)
    visibility[i*4 : (i+1)*4] = [True, True, True, True]
    buttons.append(dict(label=str(student), method="update", args=[{"visible": visibility}]))

# 5. Layout Styling - Moving the Legend Lower
fig_svm.update_layout(
    title_text="Unified Mastery Dashboard: SVM Projection & Causal Weights",
    updatemenus=[dict(
        active=0, 
        buttons=buttons, 
        x=1.1, 
        y=1.15,
        xanchor="left"
    )],
    template="plotly_white",
    height=750, # Increased height to give the layout more "breathing room"
    
    # Legend Positioning: Lowered and centered under the line chart
    legend=dict(
        orientation="h", 
        yanchor="top", 
        y=-0.25,      # Lowered further from -0.2
        xanchor="center", 
        x=0.35,
        bgcolor="rgba(255, 255, 255, 0.5)" # Semi-transparent background
    ),
    
    # Bottom margin increased to accommodate the lowered legend
    margin=dict(r=100, b=180, t=100) 
)

# Axis Labels
fig_svm.update_xaxes(title_text="Week Number", row=1, col=1)
fig_svm.update_yaxes(title_text="Performance Index", row=1, col=1)
fig_svm.update_xaxes(title_text="Weight (Impact Amount)", row=1, col=2)

fig_svm.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# 1. Prepare Data & Train Model (Lags 1-4)
pt1_ml = pt1.copy()
for i in range(1, 5):
    pt1_ml[f'perf_lag_{i}'] = pt1_ml.groupby('student_id')['performance_index'].shift(i)

pt1_ml = pt1_ml.dropna()

features = [
    'study_hours', 'sleep_hours', 'stress_level', 
    'attendance_rate', 'learning_efficiency', 'fatigue_index',
    'perf_lag_1', 'perf_lag_2', 'perf_lag_3', 'perf_lag_4'
]
X = pt1_ml[features]
y = pt1_ml['performance_index']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
model = SVR(kernel='linear', C=1.0) 
model.fit(X_scaled, y)
weights = model.coef_[0]

# 2. Setup Subplots
fig = make_subplots(
    rows=1, cols=2, 
    column_widths=[0.7, 0.3],
    subplot_titles=("Performance Projections & Baseline", "AI Impact Factors"),
    horizontal_spacing=0.12
)

student_list = pt1_ml['student_id'].unique()

# 3. Add Traces (4 traces per student)
for student in student_list:
    s_df = pt1_ml[pt1_ml['student_id'] == student]
    s_X_scaled = scaler.transform(s_df[features])
    y_pred = model.predict(s_X_scaled)
    
    # --- GLOBAL SCENARIO IMPROVEMENT ---
    s_opt = s_df[features].copy()
    
    # Behavioral Improvements
    s_opt['study_hours'] += 2.0
    s_opt['sleep_hours'] += 3.0
    s_opt['attendance_rate'] = (s_opt['attendance_rate'] + 0.1).clip(0, 1.0)
    
    # Cognitive State Improvements
    s_opt['learning_efficiency'] = (s_opt['learning_efficiency'] + 0.15).clip(0, 1.0)
    s_opt['stress_level'] *= 0.8
    s_opt['fatigue_index'] = (s_opt['fatigue_index'] - 2).clip(0, 10)
    
    # Temporal Momentum Improvement (The "Consistency" Factor)
    # We simulate a 10% boost in historical performance stability
    for i in range(1, 5):
        s_opt[f'perf_lag_{i}'] = s_opt[f'perf_lag_{i}'] * 1.05
    
    y_opt = model.predict(scaler.transform(s_opt))

    # --- Trend Chart (Col 1) ---
    fig.add_trace(go.Scatter(
        x=s_df['week'], y=s_df['performance_index'],
        name="Actual History", mode='lines+markers', visible=False,
        line=dict(color='#1e293b', width=2), legendgroup="group1"
    ), row=1, col=1)

    fig.add_trace(go.Scatter(
        x=s_df['week'], y=y_pred,
        name="AI Baseline (Model Fit)", mode='lines', visible=False,
        line=dict(color='#3b82f6', dash='dash', width=2), legendgroup="group1"
    ), row=1, col=1)

    fig.add_trace(go.Scatter(
        x=s_df['week'], y=y_opt,
        name="Projected (Full Habit Optimization)", mode='lines', visible=False,
        line=dict(color='#10b981', width=4), legendgroup="group1"
    ), row=1, col=1)

    # --- Reasoning Bar Chart (Col 2) ---
    sorted_idx = np.argsort(weights)
    fig.add_trace(go.Bar(
        x=weights[sorted_idx],
        y=[features[i] for i in sorted_idx],
        orientation='h', name="Impact Weight", visible=False,
        marker=dict(color=['#ef4444' if w < 0 else '#3b82f6' for w in weights[sorted_idx]]),
        showlegend=False
    ), row=1, col=2)

# 4. Interactivity Logic
for i in range(4): fig.data[i].visible = True

buttons = []
for i, student in enumerate(student_list):
    visibility = [False] * len(fig.data)
    visibility[i*4 : (i+1)*4] = [True, True, True, True]
    buttons.append(dict(label=str(student), method="update", args=[{"visible": visibility}]))

# 5. Final Layout with Lowered Legend
fig.update_layout(
    title_text="Student Mastery Twin: Holistic Optimization Dashboard",
    updatemenus=[dict(active=0, buttons=buttons, x=1.1, y=1.15)],
    template="plotly_white",
    height=750,
    legend=dict(orientation="h", yanchor="top", y=-0.25, xanchor="center", x=0.35),
    margin=dict(r=100, b=200, t=100)
)

# 6. Adding the "Improvement Amounts" Annotation
fig.add_annotation(
    dict(
        x=0.35,
        y=-0.25, # Pushed way down below the legend
        showarrow=False,
        text=(
            "<b>PROJECTION PARAMETERS (The 'Green Line' Logic):</b><br>"
            "• <b>Study:</b> +2.0 hrs/week  |  <b>Sleep:</b> +3.0 hr/night  |  <b>Attendance:</b> +10%<br>"
            "• <b>Efficiency:</b> +15% boost  |  <b>Stress:</b> 20% reduction  |  <b>Fatigue:</b> -2.0 index pts<br>"
            "• <b>Momentum:</b> 5% compounded improvement across all historical Lags (1-4)"
        ),
        xref="paper",
        yref="paper",
        align="left",
        bgcolor="#f0fdf4", # Light green background to match the projection line
        bordercolor="#16a34a",
        borderwidth=2,
        font=dict(size=12, color="#166534")
    )
)

fig.update_xaxes(title_text="Week Number", row=1, col=1)
fig.update_yaxes(title_text="Performance Index", row=1, col=1)
fig.update_xaxes(title_text="Influence (Negative ← | → Positive)", row=1, col=2)

fig.show()
```
